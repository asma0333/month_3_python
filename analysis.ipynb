```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# load dataset
df = pd.read_csv("hospital_data.csv")

print(df.head())

print(df.info())

print(df.describe())

# missing values
print(df.isnull().sum())
```

```python
df.iloc[:,1].value_counts().head(10).plot(kind='bar')
plt.title("Top Categories")
plt.show()
```

```python
df.hist(figsize=(8,6))
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])

sns.heatmap(numeric_df.corr(), annot=True)
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])

numeric_df.iloc[:,0].plot()
plt.title("Trend")
plt.show()
```
